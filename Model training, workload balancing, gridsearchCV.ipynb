workload balanacing

```python
# Check the original DataFrame used to create X_combined
print(X_combined.shape)
print(df.shape)
```

```python
df['predicted_priority']=xgb_model.predict(X_combined)
```

```python
df.head()
```

```python
from collections import defaultdict

#Getting all users
users = df['assigned to'].unique()

# Initializing a workload tracker
workload = defaultdict(int)


df['balanced_assigned_to'] = None

#sorting
df_sorted = df.sort_values(by='predicted_priority', ascending=False)

# Assigning tasks in round-robin based on current workload
for idx, row in df_sorted.iterrows():
    # Finding user with minimum workload
    assigned_user = min(users, key=lambda u: workload[u])
    
    # Assigning task
    df.at[idx, 'balanced_assigned_to'] = assigned_user
    workload[assigned_user] += 1
```
